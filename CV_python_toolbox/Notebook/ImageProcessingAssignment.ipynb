## Histogram Matching
In this section, I will try to use histogram matching method to each preprocessed images, using the original image as model.

```python
model_img = cv2.imread("../../homework3/result_lena.bmp")
img = cv2.imread("../../homework3/project3/lena2.bmp",0)
equ = cv2.equalizeHist(img)
result_name = "../../homework3/test.bmp"
arr = [0.1,0.4,0.3,0.3,0.2,0.1,0.01,0.01,0]
array = tool.createhisto(arr)
plt.plot(array, color = 'b')
plt.figure()
cdf = tool.calcdf(model_img)
res = np.zeros((512, 512, 3), dtype =np.uint8)
res = cdf[equ]
plt.plot(cdf, color = 'b')
plt.hist(model_img.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
plt.legend(('cdf','histogram'), loc = 'upper left')
plt.show()
plt.plot(cdf, color = 'b')
plt.xlim([0,256])
plt.legend(('cdf','histogram'), loc = 'upper left')
plt.show()
mat = np.hstack((img,equ, res))
cv2.imwrite(result_name, mat)        
im = Image.open(result_name)        
plt.imshow(im)
plt.figure()
plt.title(u"original histogram")
plt.hist(img.ravel(), 256, [0,256])
plt.show()
plt.title(u"new histogram")
plt.hist(res.ravel(), 256, [0,256])
plt.show()
```

## Image Segmentation
In this section, we will use histogram transformation for image segmentation.

```python
result_name = "../../homework3/test.bmp"
img = cv2.imread("../../homework3/project3/lena.bmp",0)
T = 40
color = np.linspace(0,255,256)
hist , bins = np.histogram(img.flatten(), 256,[0,256])
print("image mean value is", T)
while(1):
    T1 = (hist[:T]*color[:T]).sum()/hist[:T].sum()
    T2 = (hist[T:]*color[T:]).sum()/hist[T:].sum()
    temp = T
    T = int((T1+T2)/2)
    print("T1",T1,"T2",T2,"T",T)
    if abs(temp-T)<0.01:
        break
        
img1 = zeros((shape(img)[0],shape(img)[1],3),dtype = uint8)
img2 = zeros((shape(img)[0],shape(img)[1],3),dtype = uint8)
for i in range(shape(img)[0]):
    for j in range(shape(img)[1]):
        if img[i,j]<T:
            img[i,j] = 0
        else:
            img[i,j] = 255

cv2.imwrite(result_name, img)        
im = Image.open(result_name)        
plt.imshow(im)
```
